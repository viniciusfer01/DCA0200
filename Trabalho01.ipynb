# Naive - Bayes

O algoritmo Naïve-Bayes é um dos algoritmos de classificação mais populares em Machine learning, especialmente em tarefas que envolvam o processamento de texto. Ele é baseado na aplicação do teorema de Bayes, que é um teorema estatístico que descreve a probabilidade condicional de um evento, dada a ocorrência de outro evento relacionado.

O Naïve-Bayes é um classificador probabilístico, o que significa que ele estima a probabilidade de uma instância pertencer a uma classe específica. Ele usa uma abordagem de aprendizado supervisionado, ou seja, utiliza um conjunto de dados de treinamento com instâncias rotuladas para aprender a relação entre os atributos e as classes.

O algoritmo de machine learning assume que os atributos são independentes entre si, o que significa que a presença ou ausência de um atributo não afeta a presença ou ausência de outro atributo. Essa é uma simplificação forte e, portanto, o algoritmo é chamado de "naïve" (ingênuo) porque não leva em consideração a correlação entre os atributos.

A classificação com Naïve-Bayes é feita através do cálculo da probabilidade de cada classe para a instância de entrada, dado os valores dos atributos. A classe com a maior probabilidade é escolhida como a classe de saída. O cálculo da probabilidade é baseado na probabilidade a priori da classe (a probabilidade de uma instância pertencer a essa classe sem considerar os valores dos atributos) e a probabilidade condicional dos atributos dado a classe.

O Naïve-Bayes tem várias vantagens, como a sua simplicidade, rapidez de treinamento e boa performance em conjuntos de dados com muitos atributos. Ele também é robusto em relação a dados ausentes e pode ser usado em problemas com várias classes. No entanto, a suposição de independência dos atributos pode ser uma desvantagem em alguns casos, pois pode não ser realista em muitos conjuntos de dados.

Além disso, O Naïve-Bayes tem várias aplicações em aprendizado de máquina, especialmente em tarefas de classificação de texto, como a classificação de spam em e-mails, análise de sentimento em redes sociais e categorização de documentos. Ele também é usado em aplicações de reconhecimento de padrões, como a detecção de objetos em imagens e o reconhecimento de fala.

Em resumo, o algoritmo Naïve-Bayes é um classificador probabilístico simples e eficaz que é amplamente utilizado em aprendizado de máquina. Embora seja baseado em uma suposição forte de independência dos atributos, ele ainda é capaz de fornecer boas estimativas de probabilidade e resultados precisos em muitas aplicações.

## Separando os emails

80% serão destinados para treinamento e 20% para teste

```python
# Randomizando o dataset
data_randomized = sms_spam.sample(frac=1, random_state=1)

# Calculando um index aleatório para o split
training_test_index = round(len(data_randomized) * 0.8)

# Separando em treinamento e teste
training_set = data_randomized[:training_test_index].reset_index(drop=True)
test_set = data_randomized[training_test_index:].reset_index(drop=True)

print(training_set.shape)
print(test_set.shape)
```

## Limpeza dos dados

```python
training_set['SMS'] = training_set['SMS'].str.replace('\W', ' ') # Removendo pontuações
training_set['SMS'] = training_set['SMS'].str.lower()
```

```python
# Criando um vocabulário
training_set['SMS'] = training_set['SMS'].str.split()

vocabulary = []
for sms in training_set['SMS']:
   for word in sms:
      vocabulary.append(word)

vocabulary = list(set(vocabulary))
```

```python
len(vocabulary)
```

```python
word_counts_per_sms = {'secret': [2,1,1],
                       'prize': [2,0,1],
                       'claim': [1,0,1],
                       'now': [1,0,1],
                       'coming': [0,1,0],
                       'to': [0,1,0],
                       'my': [0,1,0],
                       'party': [0,1,0],
                       'winner': [0,0,1]
                      }

word_counts = pd.DataFrame(word_counts_per_sms)
word_counts.head()
```

```python
word_counts_per_sms = {unique_word: [0] * len(training_set['SMS']) for unique_word in vocabulary}

for index, sms in enumerate(training_set['SMS']):
   for word in sms:
      word_counts_per_sms[word][index] += 1
```

```python
word_counts = pd.DataFrame(word_counts_per_sms)
word_counts.head()
```

```python
training_set_clean = pd.concat([training_set, word_counts], axis=1)
training_set_clean.head()
```

```python
# Isolando spam e ham
spam_messages = training_set_clean[training_set_clean['Label'] == 'spam']
ham_messages = training_set_clean[training_set_clean['Label'] == 'ham']

# P(Spam) and P(Ham)
p_spam = len(spam_messages) / len(training_set_clean)
p_ham = len(ham_messages) / len(training_set_clean)

# N_Spam
n_words_per_spam_message = spam_messages['SMS'].apply(len)
n_spam = n_words_per_spam_message.sum()

# N_Ham
n_words_per_ham_message = ham_messages['SMS'].apply(len)
n_ham = n_words_per_ham_message.sum()

# N_Vocabulary
n_vocabulary = len(vocabulary)

# Laplace smoothing
alpha = 1
```

```python
# Inicialização dos vocabulários
parameters_spam = {unique_word:0 for unique_word in vocabulary}
parameters_ham = {unique_word:0 for unique_word in vocabulary}

# Cálculo dos parâmetros
for word in vocabulary:
   n_word_given_spam = spam_messages[word].sum()
   p_word_given_spam = (n_word_given_spam + alpha) / (n_spam + alpha*n_vocabulary)
   parameters_spam[word] = p_word_given_spam

   n_word_given_ham = ham_messages[word].sum()
   p_word_given_ham = (n_word_given_ham + alpha) / (n_ham + alpha*n_vocabulary)
   parameters_ham[word] = p_word_given_ham
```

```python
import re

def classify(message):
   '''
   message: a string
   '''

   message = re.sub('\W', ' ', message)
   message = message.lower().split()

   p_spam_given_message = p_spam
   p_ham_given_message = p_ham

   for word in message:
      if word in parameters_spam:
         p_spam_given_message *= parameters_spam[word]

      if word in parameters_ham: 
         p_ham_given_message *= parameters_ham[word]

   print('P(Spam|message):', p_spam_given_message)
   print('P(Ham|message):', p_ham_given_message)

   if p_ham_given_message > p_spam_given_message:
      print('Label: Ham')
   elif p_ham_given_message <= p_spam_given_message:
      print('Label: Spam')
```

```python
classify('WINNER!! This is the secret code to unlock the money: C3421.')
```

```python
classify("Sounds good, Tom, then see u there")
```

## Teste do modelo

```python
def classify_test_set(message):
   '''
   message: a string
   '''

   message = re.sub('\W', ' ', message)
   message = message.lower().split()

   p_spam_given_message = p_spam
   p_ham_given_message = p_ham

   for word in message:
      if word in parameters_spam:
         p_spam_given_message *= parameters_spam[word]

      if word in parameters_ham:
         p_ham_given_message *= parameters_ham[word]

   if p_ham_given_message > p_spam_given_message:
      return 'ham'
   elif p_spam_given_message >= p_ham_given_message:
      return 'spam'
```

```python
test_set['predicted'] = test_set['SMS'].apply(classify_test_set)
test_set.head()
```

```python
correct = 0
total = test_set.shape[0]

for row in test_set.iterrows():
   row = row[1]
   if row['Label'] == row['predicted']:
      correct += 1

print('Correct:', correct)
print('Incorrect:', total - correct)
print('Accuracy:', correct/total)
```
